```python
import matplotlib.pyplot as plt
import numpy as np

def generate_random_signal(num_points=100):
    # Generate random x values (e.g., time steps)
    x = np.linspace(0, 10, num_points)

    # Generate random y values (random signal)
    y = np.random.randn(num_points)

    return x, y

def plot_random_signal(x, y):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Random Signal', color='b')
    #plt.title('Random Signal')
    #plt.xlabel('Time')
    #plt.ylabel('Amplitude')
    #plt.legend()
    plt.grid(True)
    
    # Remove ticks and labels on both axes
    plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False,
                    labelbottom=False, labelleft=False)
    
    plt.show()

# Generate and plot random signal
x_values, y_values = generate_random_signal()
plot_random_signal(x_values, y_values)
```
